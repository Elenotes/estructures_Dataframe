```python
#the uniqueCarrier with higher delays is NW
new_df[new_df["DepDelay"] > 0].sort_values(["DepDelay", "UniqueCarrier"], ascending = False).head(15)
```

> Creem una nova columna amb la distancia de milles a km, suposant que les distancies són donades en milles.

```python
# We create a new column with distance in kilometers
# 1 mile = 1,609 km
new_df["distancia_km"] = new_df["Distance"] * 1.609
new_df.head(10)
#new_df.drop("distancia_m", inplace=True, axis=1)
```

```python
# create a new column with the flight speed --> v = km/h 
#"ActualElapsedTime" are the flight time in minutes
temps_hores = new_df["ActualElapsedTime"] / 60

new_df["velocitat"] = new_df["distancia_km"] / temps_hores
new_df.head(10)
#new_df[new_df["velocitat"] > 1000]
```

> TOTS els vols (100%) tenen algun retràs en l'hora de sortida. Mentre que en l'hora d'arribada són el 89.37%

```python
#percent of delayed flights
new_df[new_df["DepDelay"] > 0].count()/len(new_df) * 100
new_df[new_df["ArrDelay"] > 0].count()/len(new_df) * 100
```

```python
new_df["distancia_km"].sort_values(ascending = False)
```

```python
#new_df["Dest"].unique()
#calculate the means of ArrDelay, DepDelay, Distance_km...
mitjana_arrDelay = new_df["ArrDelay"].mean()
mitjana_depDelay = new_df["DepDelay"].mean()
mitjana_distancia = new_df["distancia_km"].mean()
velocitat_mitjana = new_df["velocitat"].mean()

print("Algunes mitjanes generals:")
print("**************************")
print("Mitjana de retràs en les sortides: {:.2f} minuts.".format(mitjana_depDelay))
print("Mitjana de retràs en les arribades: {:.2f} minuts.".format(mitjana_arrDelay))
print("Distància mitjana recorreguda pels avions durant el 2008: {:.2f} km.".format(mitjana_distancia))
print("Velocitat mitjana dels avions durant el 2008: {:.2f} km/h.".format(velocitat_mitjana))
```

> Treballem el **groupby()**

> La companyia amb una mitjana de retrassos més alta és la **"YV"**, amb 55 minuts.
    La que acomula menys retrassos és la **"AQ"** amb 21 minuts. Hi ha 10 companyies que acomules una mitjana de més de 3/4 d'hora de retràs.
    
> La companyia amb més vols **"WN"**, és la tercera que acomula menys retrassos.

```python
#Arrivals delay means by UniqueCarrier
grouped = new_df.groupby('UniqueCarrier')
grouped['ArrDelay'].agg(np.mean).sort_values(ascending = False)
```

> **agg()** method. Podem veure que les dades màximes i mínimes tenen valors estranys, amb la qual cosa, s'haurien de netejar abans de fer-ne les estadístiques.

```python
new_df.agg(
    {"ArrDelay": ["min", "max", "median"],
     "DepDelay": ["min", "max", "median"],
     "velocitat": ["min", "max", "median"],
     "distancia_km":["min", "max", "median"]
    }
)
```
